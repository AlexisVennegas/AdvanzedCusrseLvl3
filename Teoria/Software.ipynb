# Multiusuario

Un sistema operativo multiusuario es un sistema que puede dar servicio a varios usuarios de
forma simultánea. Hay que tener en cuenta que un sistema con varias cuentas de usuario no
es necesariamente un sistema multiusuario. Por ejemplo, las versiones domésticas de los
sistemas Windows® permiten tener varias cuentas de usuario en el mismo sistema operativo
pero no permiten que haya varios usuarios trabajando en el sistema al mismo tiempo. Las
versiones Server de Windows® sí permiten esta característica a través de terminal server.
Por el contrario, las versiones Linux sí son multiusuario desde hace mucho tiempo.

- Centralizados. En este modelo un ordenador central se encarga de todo el
procesamiento y los usuarios se conectan a él a través de terminales que carecen de
memoria y procesador.
- Distribuidos. Permiten distribuir los trabajos, tareas o procesos entre un conjunto de
procesadores, que pueden estar en el mismo equipo o en equipos distintos.
- De escritorio. Es el utilizado habitualmente por los equipos de sobremesa, estaciones
de trabajo y portátiles
- ● En red. Son los que permiten mantener unidad entre dos o más ordenadores a través
de algún medio de comunicación para poder compartir los recursos y la información
del sistema.

# Principales Funciones de los sistemas operativos


- Políticas de reparto de tiempo de proceso. Controla la ejecución de varios procesos a
la vez, repartiendo los recursos del ordenador (procesador, memoria, espacio de
almacenamiento…) a los distintos programas que se están ejecutando.
- Control y manejo de los dispositivos de E/S. Controla y organiza los dispositivos
conectados al sistema.
- Control de recursos. Coordinar y manipular el hardware del sistema informático. Se
encarga del funcionamiento coordinado de todos los componentes para que funcionen
como una sola máquina.
- Administra y mantiene los sistemas de archivo de disco permitiendo guardar la
información en las unidades de almacenamiento en forma de ficheros y directorios.
- Gestiona y mantiene usuarios y grupos de usuarios pudiendo establecer privilegios y
restricciones a las diferentes cuentas individuales o grupales.
- Ofrece una base estándar sobre la que ejecutar otros programas permitiendo diseñar
software de aplicación sin necesidad de tener en cuenta el hardware particular de
cada sistema.
- Permite la interacción entre el sistema y los usuarios permitiendo su manejo de forma
fácil e intuitiva a través de la interfaz gráfica o GUI.
- Detecta e informa de los errores que se produzcan.

# Políticas de reparto de tiempo de Proceso

### Algunos algoritmos de planificación

- FIFO o FCFS (First Come First Served): Atiende a cada proceso completamente en su turno, uno después del otro, hasta terminar.
- Round Robin: Da a cada proceso un poquito de tiempo, luego pasa al siguiente y vuelve a dar otra oportunidad más adelante, y así en ciclos.
- SJF: Atiende primero las tareas más cortas, para reducir la espera de las tareas rápidas.
- Planificación con múltiples colas: ivide las tareas en varias filas según su tipo o prioridad, y organiza la atención según la importancia de cada fila.
- Planificación por prioridades: Atiende primero las tareas más importantes o urgentes, sin importar el tiempo de espera o de ejecución.

El sistema de entrada/salida es aquella parte del sistema operativo encargada de abstraerse
los dispositivos de entrada/salida conectados al equipo y ofrecer una visión más simplificada
de los mismos de tal forma que los demás componentes del SSOO o software puedan hacer
uso de ellos de una manera fácil.
 Los dispositivos de entrada/salida se pueden clasificar en tres grupos:
- Dispositivos de almacenamiento. Como su nombre indica, van a servir al sistema
para almacenar la información del mismo de forma no volátil. Son dispositivos de este

- Dispositivos de interfaz de usuario. Permiten la interacción entre el usuario y el
equipo. Ratones, teclado,pantalla, impresoras, etc., son dispositivos de este tipo.

# Capítulo II Sistemas de archivo

## Implementación de sistemas de archivo


 se refiere al diseño y construcción del software que permite a un sistema operativo organizar, almacenar, recuperar y gestionar los datos en dispositivos de almacenamiento, como discos duros, SSDs o unidades flash.

## Capas de abstracción

En un sistema de archivos, las **capas de abstracción** organizan el flujo de datos desde las aplicaciones hasta el hardware de almacenamiento. Esto permite manejar archivos de manera eficiente y mantener la integridad y seguridad de los datos. Estas capas típicas son:

### 1. **Capa de Aplicación**
   - Es la capa más alta, donde interactúan los programas o aplicaciones del usuario con el sistema de archivos, a través de operaciones como abrir, leer, escribir, o cerrar archivos.
   - A través de esta capa, las aplicaciones hacen solicitudes de E/S (entrada/salida) al sistema de archivos, sin preocuparse por los detalles de almacenamiento físico.

### 2. **Capa del Sistema de Archivos Lógico**
   - Esta capa organiza y gestiona la estructura lógica de los datos y metadatos de los archivos, como nombres de archivos, permisos, estructura de directorios y ubicaciones.
   - Incluye la gestión de permisos de acceso, los sistemas de control de usuarios, y los aspectos lógicos de la organización de archivos y carpetas.
   - Aquí, el sistema de archivos traduce las operaciones de archivos de alto nivel a bloques de datos, manejando aspectos como la fragmentación.

### 3. **Capa de Gestión de Archivos**
   - En esta capa se maneja la asignación de bloques a los archivos, y se realiza el seguimiento de los bloques libres y ocupados en el disco.
   - Es responsable de administrar los espacios libres y asignar almacenamiento a los archivos según las solicitudes de la capa superior.
   - Esta capa puede incluir métodos como el uso de tablas (por ejemplo, FAT), índices (como en NTFS), o estructuras de árboles (como en ext4) para localizar archivos.

### 4. **Capa de Sistema de Entrada/Salida (E/S) de Archivos**
   - Esta capa traduce las operaciones de archivos a operaciones de bloques de datos en el almacenamiento físico.
   - A través de esta capa, el sistema de archivos interactúa con los controladores de dispositivos para escribir y leer datos en bloques específicos.
   - También incluye el manejo de búferes y cachés para optimizar el rendimiento de lectura/escritura y reducir el tiempo de espera de acceso a los datos.

### 5. **Capa del Controlador de Dispositivos**
   - Esta capa actúa como intermediario entre el sistema operativo y el hardware de almacenamiento.
   - Aquí se encuentran los controladores de hardware específicos que traducen las operaciones genéricas de E/S en comandos específicos del dispositivo.
   - Los controladores también gestionan detalles como la sincronización de operaciones y el manejo de interrupciones de hardware.

### 6. **Capa Física (Hardware)**
   - Es la capa más baja, donde se almacenan físicamente los datos en los discos duros, SSDs, o cualquier dispositivo de almacenamiento.
   - Esta capa define cómo se estructuran los datos a nivel físico, manejando la posición en el disco o memoria y controlando el acceso eléctrico o magnético a los datos.

### Resumen de las Capas

| Capa                    | Función Principal                                                  |
|-------------------------|--------------------------------------------------------------------|
| **Aplicación**          | Interacción de programas con archivos (abrir, cerrar, leer, etc.) |
| **Sistema de Archivos Lógico** | Organización lógica de archivos y directorios                |
| **Gestión de Archivos** | Asignación y gestión de bloques de datos                           |
| **Sistema de E/S de Archivos** | Traducción a operaciones de bloques en disco               |
| **Controlador de Dispositivos** | Comunicación directa con el hardware                      |
| **Física (Hardware)**   | Almacenamiento físico de datos                                    |

Estas capas de abstracción permiten que el sistema de archivos maneje de forma flexible y eficiente la gestión de datos en los dispositivos, sin que las aplicaciones necesiten conocer la complejidad del almacenamiento físico.

## Comparación de sistemas de archivos


- NTFS: Sistema de Windows con funciones avanzadas (compresión, cifrado) pero ocupa mucho espacio.
- HFS+: Usado por Apple, permite cifrado pero no compresión.
- EXT4: Común en Linux, es rápido y eficiente, sin compresión ni cifrado, pero compatible con EXT3.

## Capítulo III Multiproceso, Multitarea y MultiUsuario


## Multiproceso
Un sistema operativo multiproceso es aquel que puede ejecutar varios procesos de forma concurrente (a la vez). Para que se puedan ejecutar varios procesos a la vez, es necesario tener un sistema multiprocesador (con varios procesadores).

El objetivo de utilizar un sistema con varios procesadores es aumentar la potencia de cálculo y, por lo tanto, el rendimiento. El sistema reparte la carga de trabajo entre los procesadores y gestiona la memoria para distribuirla adecuadamente. Los sistemas multiprocesador suelen usarse en estaciones de trabajo, servidores, etc.

<!-- ## Multitarea -->
En un sistema operativo multitarea, varios procesos pueden ejecutarse aparentemente al mismo tiempo sin que el usuario lo perciba. La mayoría de los sistemas operativos actuales son multitarea.

Un sistema multitarea permite, por ejemplo, escuchar música, navegar por Internet y realizar una videoconferencia simultáneamente. El sistema operativo fracciona el tiempo de CPU y lo distribuye entre los procesos que lo requieren de la mejor forma posible.

## Multihilo
Un proceso puede tener varios hilos de ejecución, permitiendo que un proceso ejecute varias tareas al mismo tiempo. El multihilo es eficiente, ya que crear muchos procesos requiere más recursos, mientras que varios hilos pueden compartir los recursos y la memoria de un proceso. Además, el multihilo evita la pérdida de tiempo en cambios de contexto, ya que todos los hilos se ejecutan en el mismo contexto.

## Multiusuario
Un sistema operativo multiusuario puede dar servicio a varios usuarios simultáneamente. No hay que confundir un sistema con varias cuentas de usuario con un sistema multiusuario.

Por ejemplo, las versiones domésticas de Windows® permiten tener varias cuentas de usuario, pero no permiten que varios usuarios trabajen al mismo tiempo. En cambio, las versiones Server de Windows® sí lo permiten a través de terminal server. Las versiones de Linux han sido multiusuario desde hace mucho tiempo.


# Capítulo IV Particionamiento Lógico y Núcleos Virtuales



## Historia de la Virtualización
La virtualización tiene sus inicios en los años 60, cuando IBM desarrolló sistemas de particiones lógicas para sus mainframes, permitiendo que varias máquinas virtuales compartieran un solo sistema físico. Con la llegada de la arquitectura x86 en los 80, el modelo cambió y se redujo la necesidad de virtualización debido al uso de múltiples máquinas físicas.

En los 90, la virtualización resurgió como una solución más económica y eficiente para simplificar sistemas, reducir costos y facilitar la administración. Al principio, los sistemas de virtualización en x86 tenían un rendimiento bajo, pero evolucionaron hasta permitir sistemas operativos virtuales de tipo servidor y la creación de hypervisores. Actualmente, la virtualización es esencial en entornos empresariales, optimizando el uso de hardware, consumo de energía y gestión.

## Concepto de Virtualización
La virtualización se implementa en una máquina anfitriona mediante una capa de software que asigna recursos como procesador, memoria y almacenamiento. Esta capa permite instalar y ejecutar sistemas operativos huéspedes que utilizan recursos virtuales, gestionados a través de drivers virtuales, para interactuar con las aplicaciones y el sistema operativo.

La virtualización moderna ha mejorado significativamente gracias al soporte nativo de hardware, lo que la hace más eficiente y sencilla de administrar.

# Sistemas de Archivos

## Sistemas de archivos de disco
Estos sistemas permiten que múltiples procesos accedan a un fichero sin importar su ubicación en el disco, optimizando el acceso y anticipándose a las demandas de los procesos. Los sistemas más comunes son:

- **FAT16 (FAT)**: Particiones de hasta 2 GB, archivos de máximo 2 GB, unidades de asignación de 32 KB, no distingue mayúsculas/minúsculas. Compatible con la mayoría de los sistemas operativos.
- **FAT32**: Archivos de hasta 4 GB y particiones de más de 2 GB, con unidades de asignación de 4 KB. Máximo tamaño de partición de 2 TB. No distingue entre mayúsculas/minúsculas.
- **NTFS**: Volúmenes de hasta 16 EB (implementación actual 256 TB), distingue mayúsculas/minúsculas. Unidades de asignación entre 4 KB y 64 KB.
- **EXT3**: Propio de Linux, permite particiones de hasta 32 TB y archivos de hasta 2 TB. Distingue mayúsculas/minúsculas.
- **EXT4**: Utilizado en Linux, soporta particiones de hasta 1 EB y archivos de 16 TB. Distingue mayúsculas/minúsculas.
- **HFS+**: Sistema de Apple, particiones de hasta 16 EB y archivos de hasta 8 EB.

## Sistemas de archivos de discos ópticos
Utilizados en CD, DVD y discos Blu-ray. Los más comunes son:

- **ISO 9660**
- **UDF (Universal Disk Format)**

## Sistemas de archivo de cintas
Debido al almacenamiento secuencial, los accesos aleatorios son más lentos que en discos tradicionales, ya que requieren rebobinar la cinta.

## Sistemas de archivo de red
Distribuidos en una red local, donde los clientes acceden a los archivos proporcionados por un servidor. Utilizan protocolos como:

- **NFS (Network File System)**
- **SMB (Server Message Block)**

### Tipos de Particiones en un Disco Duro

En un disco duro, las particiones permiten dividir el espacio de almacenamiento en secciones independientes para distintos usos. Los tipos de particiones más comunes son:

#### 1. Primaria
- **Uso principal**: Se utilizan principalmente para instalar sistemas operativos, aunque estos también pueden instalarse en particiones lógicas.
- **Limitación**: Un disco puede tener hasta **4 particiones primarias** como máximo.
- **Ejemplo de uso**: Instalación de un sistema operativo o software que requiera acceso directo a una partición primaria.

#### 2. Extendida
- **Propósito**: Una partición extendida no contiene datos directamente, sino que sirve para albergar **particiones lógicas**.
- **Características**: Solo puede existir **una partición extendida** por disco, pero puede contener múltiples particiones lógicas dentro de ella.
- **Ventaja**: Permite superar la limitación de 4 particiones en un disco, facilitando la creación de más particiones para diversos usos.

#### 3. Lógica
- **Definición**: Son particiones “secundarias” que se encuentran dentro de una partición extendida.
- **Uso común**: Se utilizan principalmente para almacenar datos, programas o archivos personales.
- **Flexibilidad**: No hay un límite estricto en el número de particiones lógicas que se pueden crear dentro de una partición extendida, lo que ofrece mayor flexibilidad para gestionar el almacenamiento.

### Resumen
Las particiones primarias son ideales para sistemas operativos y tareas críticas, las extendidas permiten la creación de particiones lógicas adicionales, y las lógicas son perfectas para almacenamiento secundario y organización de datos.


### El sector de arranque
Un disco se compone de un sector de arranque y una serie de particiones y opcionalmente
espacio sin particionar.
El sector de arranque es el primer sector del disco (cabeza 0, cilindro 0 y sector 1).
Dentro de él está la tabla de particiones y el Master Boot(MBR) o gestor de arranque.
Este programa lee la tabla de particiones y cede el control al sector de arranque de la partición
activa. Como se ha dicho antes si no hay partición activa, el equipo da un error al arrancar.

## NAS (Network Attached Storage)
Un NAS es un dispositivo de almacenamiento conectado a una red que permite a los usuarios almacenar y acceder a datos a través de una red, como si fuera una unidad de disco externa, pero con la ventaja de que se puede acceder de manera remota desde varios dispositivos y sistemas operativos.

Características:
Almacenamiento de archivos: El NAS está diseñado para almacenar y compartir archivos a través de una red.
Acceso a través de protocolo de red: Generalmente utiliza protocolos como SMB (Server Message Block) o NFS (Network File System) para compartir archivos entre dispositivos en la red.
Fácil de configurar: Suele ser más fácil de configurar en comparación con un SAN y es ideal para pequeñas empresas o usuarios domésticos.
Escalabilidad limitada: Aunque algunos modelos permiten expandir la capacidad de almacenamiento, no está diseñado para gestionar grandes volúmenes de datos o tareas críticas a nivel empresarial.
Ejemplo:
Un hogar que tiene múltiples dispositivos como PCs, laptops, tablets y smartphones, podría usar un NAS para almacenar y compartir música, fotos y documentos. Un ejemplo de NAS sería el Synology DiskStation o QNAP TS Series.

SAN (Storage Area Network)
Una SAN es una red especializada de alto rendimiento que conecta servidores a dispositivos de almacenamiento. A diferencia del NAS, que trabaja a nivel de archivos, el SAN opera a nivel de bloques de datos, lo que permite que las aplicaciones accedan al almacenamiento como si fuera un disco local, brindando mayor velocidad y capacidad para manejar grandes volúmenes de datos.

Características:
Almacenamiento a nivel de bloques: En lugar de compartir archivos, el SAN permite a los servidores acceder a bloques de datos, lo que hace que el acceso sea más rápido y eficiente.
Alto rendimiento: Está diseñado para aplicaciones empresariales que requieren acceso rápido a grandes volúmenes de datos.
Requiere infraestructura más compleja: Configurar un SAN puede ser más complejo y costoso debido a los componentes y el hardware especializados, como switches y cables FC (Fibre Channel).
Escalabilidad: Las SANs pueden escalarse fácilmente para manejar grandes cantidades de datos y usuarios, lo que las hace ideales para grandes empresas.
Ejemplo:
Una empresa grande con múltiples servidores que necesita almacenar grandes bases de datos o archivos de aplicaciones, como un servidor de correo o un servidor de virtualización. Un ejemplo de SAN sería EMC VMAX o NetApp FAS Series.

# Protección RAID

El sistema **RAID** (Redundant Array of Independent Disks) es una tecnología que utiliza múltiples discos duros para mejorar el rendimiento, la redundancia y la disponibilidad de los datos en un sistema de almacenamiento. La **protección RAID** se refiere a la **redundancia de datos** que proporciona RAID para protegerlos contra fallos de disco.

## Tipos de RAID comunes con protección de datos:

### RAID 1 - Espejado (Mirroring)
- **Protección**: Los datos se duplican en dos discos. Si uno de los discos falla, el sistema puede seguir funcionando con el otro.
- **Ejemplo**: Si tienes dos discos de 500 GB configurados en RAID 1, los 500 GB de datos se almacenarán en ambos discos. Si uno de los discos falla, no perderás ningún dato.

### RAID 5 - Paridad distribuida
- **Protección**: Los datos y la paridad (información de recuperación) se distribuyen entre todos los discos. Si un disco falla, la información perdida se puede reconstruir utilizando la paridad.
- **Ejemplo**: En un RAID 5 con tres discos de 1 TB, la capacidad usable sería 2 TB, ya que el 1 TB restante se utiliza para la paridad. Si uno de los discos falla, se puede recuperar toda la información utilizando los datos y la paridad distribuidos en los otros discos.

### RAID 10 (1+0) - Combinación de RAID 1 y RAID 0
- **Protección**: Combina la redundancia del RAID 1 con la velocidad del RAID 0. Los datos se espejan (como en RAID 1) y luego se distribuyen entre los discos (como en RAID 0).
- **Ejemplo**: Si tienes 4 discos de 500 GB configurados en RAID 10, tendrás 1 TB de capacidad utilizable. Ofrece mayor redundancia y rendimiento, ya que puedes perder hasta 2 discos (si son del mismo par espejado) sin perder datos.

## Ventajas de la protección RAID:
- **Alta disponibilidad**: RAID garantiza que los datos estén protegidos y accesibles incluso si un disco falla.
- **Recuperación de fallos**: La paridad o duplicación de datos permite la recuperación de los mismos en caso de fallo de un disco.
- **Mejora del rendimiento**: Algunos niveles de RAID también mejoran la velocidad de lectura y escritura de los datos.

## Desventajas:
- **Requiere más discos**: Los sistemas RAID con protección de datos (como RAID 1 y RAID 5) necesitan más discos que un sistema sin RAID.
- **Riesgo de fallo de múltiples discos**: Aunque RAID ofrece protección contra fallos de un disco, puede no ser suficiente si fallan más de uno al mismo tiempo.

En resumen, RAID es una solución de almacenamiento que mejora la **seguridad** y la **disponibilidad** de los datos al repartir la carga entre varios discos y protegerlos contra fallos.

## Estrategias Principales
El plan se basa en tres estrategias clave:
- **Prevención**: Implementar medidas proactivas para evitar incidentes.
- **Mitigación**: Minimizar el impacto de los incidentes cuando ocurran.
- **Recuperación**: Restablecer operaciones en el menor tiempo posible.

Para asegurar la continuidad del negocio, se deben incluir:
1. **Copias de Seguridad**:
   - Almacenar datos en **centros de datos separados** geográficamente.

   - Utilizar enlaces rápidos entre los centros para replicar datos en tiempo real.

   - Implementar sistemas redundantes que asuman la carga en caso de fallos.

Es fundamental contar con medidas de protección física:
- **Control de acceso**: Restricción de entrada a zonas críticas.
- **Protección del cableado**: Resguardar la infraestructura de comunicaciones.
- **Precauciones antiincendio**: Detectores de humo y sistemas de extinción.
- **Control de inundaciones**: Sensores y barreras para evitar daños por agua.


# Capítulo III: Políticas de Salvaguarda

La supervivencia de una empresa depende de factores internos y externos, incluyendo económicos, demográficos y de mercado. Es responsabilidad de la empresa implementar medidas que aseguren la continuidad de sus operaciones ante situaciones adversas.

Las grandes empresas suelen contar con un **plan de continuidad de negocio**, que establece los recursos, capacidades y procedimientos necesarios para mitigar los efectos de riesgos o eventos externos. Este plan se basa en tres estrategias principales:

- **Estrategias de prevención**
- **Estrategias de mitigación**
- **Estrategias de recuperación**

Para garantizar la continuidad, es fundamental reducir el tiempo de recuperación y el punto de recuperación. Esto se logra mediante:

- **Copia de datos en centros distintos**
- **Conexiones de alta velocidad entre centros**
- **Infraestructura paralela para absorber actividad en caso de incidentes**

El plan de continuidad también contempla la seguridad física, que incluye:

- **Control de acceso a la sala**
- **Protección del cableado estructurado**
- **Precauciones antiincendio**
- **Sistemas de control ante inundaciones**

## Salvaguarda Lógica
La **salvaguarda lógica** consiste en la aplicación de barreras y procedimientos que resguarden el acceso a los datos y solo se permita acceder a ellos a las personas autorizadas para hacerlo.

### Objetivos Principales:
- **Restringir el acceso** a los programas y archivos a usuarios sin los permisos correspondientes.
- **Asegurar que los usuarios puedan trabajar sin supervisión exhaustiva** y no puedan modificar los programas ni los archivos que no les correspondan.
- **Asegurar el uso correcto** de los datos, archivos y programas por el procedimiento adecuado.
- **Controlar la transmisión de información** para que solo el destinatario previsto la reciba.
- **Comprobar la integridad** de la información transmitida y recibida.
- **Implementar sistemas alternativos** de transmisión entre diferentes puntos.
- **Disponer de soluciones de emergencia** para la transmisión de información.

Estos controles pueden implementarse en:
- **Sistemas operativos**
- **Sistemas de aplicación**
- **Bases de datos**
- **Paquetes específicos de seguridad**

Estos mecanismos ayudan a proteger el sistema operativo de la red, el sistema de aplicación y el software de la utilización o modificaciones no autorizadas, garantizando la integridad de la información y restringiendo el acceso a usuarios y procesos autorizados. Además, se debe complementar con **backups o clonaciones** para evitar la pérdida de datos por errores o fallos de hardware o software.
