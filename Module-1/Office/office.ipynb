# Capítulo V: Transferencia de Ficheros

Existen varias formas de transferir ficheros entre máquinas. A continuación, se describen las más comunes:

## Transferencia en Redes Locales
- **Protocolos**: 
  - **SMB (Server Message Block)**: Utilizado en Windows y Unix; también conocido como CIFS (Common Internet File System). Permite compartir carpetas e impresoras.
  - **NFS (Network File System)**: Popular en máquinas Unix.
- **SMB 3.0**: Presente en Windows 8 y Server 2012, mejora rendimiento y funcionalidades.
- **Samba**: Implementación de SMB/CIFS en Unix. Se configura mediante ficheros o herramientas gráficas como Swat.

## Transferencia mediante FTP
- **Protocolo**: Creado en 1971 para transferencias bajo arquitectura cliente-servidor.
- **Autenticación**: Envía credenciales sin cifrar, aunque admite acceso anónimo y versiones seguras como **FTPS** o **SFTP**.
- **Uso**: Popular en servidores web para gestionar archivos HTML. Navegadores como Firefox soportan extensiones como FireFTP para facilitar transferencias.

## Transferencia mediante HTTP
- **Características**:
  - Eficiente para ficheros pequeños.
  - Generalmente no bloqueado por firewalls (puertos 80 y 443).
  - Admite compresión de datos, agilizando la transferencia.
- **Ventajas**: Menos problemas con proxys en comparación con FTP.

## Transferencia Peer-to-Peer (P2P)
- **Redes P2P**: Sin paradigma cliente-servidor; las máquinas actúan como cliente y servidor simultáneamente.
- **Ventajas**: 
  - Aprovechan el ancho de banda de los usuarios.
  - Mayor velocidad de descarga.
- **Problemas**: Uso indebido para compartir archivos protegidos por derechos de autor.
- **Ejemplo**: Clientes como BitTorrent.



# Capítulo VI: Proyectos de Software Libre en la Red

## Software Libre
- **Definición**: Software que incluye el código fuente, permitiendo a cualquier persona:
  - Modificarlo.
  - Redistribuirlo, incluyendo las modificaciones.
  - Usarlo sin restricciones.
- **Ejemplo**: Similar a compartir la "fórmula secreta" de un producto como Coca-Cola™.
- **Beneficios**: Más control sobre el producto, posibilidad de adaptarlo o mejorarlo.
- **Ejemplo destacado**: **Linux**:
  - Creado por **Linus Torvalds** bajo licencia GNU.
  - Mejorado continuamente por programadores de todo el mundo.
- **Mitos desmentidos**: 
  - Se critica por falta de garantía y soporte técnico, pero la comunidad demuestra rapidez y eficiencia en resolver problemas.
- **Proyectos relevantes**: Mozilla, lenguaje PERL, Apache, formato PNG.

## La Licencia Pública GNU (GNU GPL)
- **Creada por**: Richard Stallman y la **Free Software Foundation (FSF)** en 1985.
- **Propósito**: Difundir y promover el software libre, además de desarrollar el sistema operativo GNU.
- **Permite**:
  - Modificar, compartir, utilizar y estudiar software distribuido bajo esta licencia.
- **Condición**: 
  - Los trabajos derivados deben distribuirse bajo la misma licencia (**copyleft**).
  - Ejemplo: Si un programa utiliza código bajo licencia GPL, todo el programa debe licenciarse bajo GPL.



# Capítulo VII: Sistemas de Control de Versiones

## ¿Qué es un Sistema de Control de Versiones?
- Permite gestionar los cambios en un producto durante su desarrollo o vida útil.
- Especialmente útil cuando participan varias personas, ya que evita errores manuales y facilita la organización.

---

## Tipos de Sistemas de Control de Versiones

### **1. Centralizados**
- **Características**:
  - Existe un repositorio único donde se almacena todo el código.
  - Las versiones o ramas requieren aprobación de un responsable.
- **Ventajas**:
  - Mayor control sobre el proyecto.
  - Número de versión único.
- **Ejemplo**: CVS.

### **2. Distribuidos**
- **Características**:
  - Cada usuario tiene un clon completo del repositorio (backup).
  - Generalmente existe un repositorio principal en el servidor.
- **Ventajas**:
  - Almacena versiones inestables o de prueba.
  - Mayor autonomía y flexibilidad para los desarrolladores.
  - Reduce la carga del servidor central.
  - Permite crear ramas para nuevas funcionalidades o reestructuraciones.
- **Ejemplo**: GIT.

---

## Formas de Colaboración

### **1. Exclusiva**
- Cada usuario se encarga de un elemento específico.
- El repositorio bloquea el elemento mientras está en desarrollo.
- Una vez completado, se actualiza el repositorio y se desbloquea.

### **2. Colaborativa**
- Cada usuario trabaja en su copia local.
- Los cambios se suben al repositorio común una vez terminados.
- **Problema**: Puede generar conflictos al no estar coordinadas las modificaciones.

---

## Flujos de Trabajo (Workflow)

### **1. Centralizado**
- Utiliza un servidor centralizado.
- Los desarrolladores publican actualizaciones en el servidor.
- Exclusivo: un elemento en desarrollo no puede ser modificado por otro desarrollador.

### **2. Con Gestor de Integración**
- Un integrante del equipo actúa como **gestor de integración**, actualizando el repositorio común.
- Los desarrolladores suben su trabajo a repositorios públicos.
- Común en SCV distribuidos.

### **3. Con Dictador y Tenientes**
- Diseñado para **proyectos masivos** (e.g., kernel de Linux).
- **Roles**:
  - **Tenientes**: Supervisan distintas partes del proyecto.
  - **Dictador**: Decide qué cambios integrar en el repositorio principal.
- Todos los desarrolladores pueden clonar y crear copias locales del repositorio.
